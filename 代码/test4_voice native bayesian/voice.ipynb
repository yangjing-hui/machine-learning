```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import pandas as pd

data = pd.read_csv('voice.csv')
data.head()
# 假设data是你的数据集，特征列为features，性别标签列为label
# data.head()应该展示数据的前几行，确保数据被正确加载
```

```python
# 随机选取训练集和测试集
train_data, test_data = train_test_split(data, train_size=2000, test_size=1168, random_state=42)

# 分离特征和标签
X_train = train_data.drop('label', axis=1)  # 特征
y_train = train_data['label']  # 标签
X_test = test_data.drop('label', axis=1)
y_test = test_data['label']

# 创建并训练朴素贝叶斯分类器
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
```

```python
# 预测测试集
y_pred = naive_bayes_classifier.predict(X_test)

# 计算准确度
accuracy = accuracy_score(y_test, y_pred)

# 输出准确度
print(f"准确度: {accuracy * 100:.2f}%")
```
